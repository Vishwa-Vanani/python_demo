```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("laptops.csv")
df
```

```python
# 1.Display information of laptops having brand “Apple” and model name “Macbook Pro”.

apple_macbook_pro = df[(df["Brand"].str.lower() == "apple") & (df["Model Name"].str.lower() == "macbook pro")]
apple_macbook_pro
```

```python
# 2.Display information of laptops having Windows Operating System.

windows_laptops = df[df['Operating System'] == 'Windows']
#windows_laptops = df[df['Operating System'].str.lower() == 'windows']
windows_laptops
```

```python
# 3.Display the information of laptop having 16GB RAM and MacOS.

# Convert RAM to numeric
#df['RAM_GB'] = df['RAM'].str.replace('GB', '').astype(int)

# macos_16GB = df[(df['RAM_GB'] == 16) & (df['Operating System'].str.lower() == 'macos')]
macos_16GB = df[(df['RAM'] == '16GB') & (df['Operating System'] == 'MacOS')]
macos_16GB
```

```python
# 4.Calculate the average price of Dell laptops.

# avg_price_of_dell_laptops = df[df['Brand'].str.lower() == 'dell']['Price'].mean()
avg_price_of_dell_laptops = df[df['Brand'] == 'Dell']['Price'].mean()
avg_price_of_dell_laptops
```

```python
# 5. Calculate the Model Name wise Apple laptop’s average price.

# model_wise_avg_price_of_apple = df[df['Brand'].str.lower() == 'apple'].groupby('Model Name')['Price'].mean()
model_wise_avg_price_of_apple = df[df['Brand'] == 'Apple'].groupby('Model Name')['Price'].mean()
model_wise_avg_price_of_apple
```

```python
# 6. Graphically show the comparative data of laptops brands.

brand_avg_price = df.groupby('Brand')['Price'].mean().sort_values(ascending = False)
# brand_avg_price
plt.figure(figsize = (8, 5))
plt.bar(brand_avg_price.index, brand_avg_price.values, color='skyblue')
plt.title("Average Price per Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price (₹)")
plt.tight_layout()
plt.show()
```

```python
# 7. Graphically show the relationship between RAM and Price of laptops.
plt.figure(figsize=(7,5))
plt.scatter(df['RAM'], df['Price'], color='purple')
plt.title("Relationship between RAM and Price")
plt.xlabel("RAM (GB)")
plt.ylabel("Price (₹)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 8. Graphically show the relationship between screen size and price of laptops.
plt.figure(figsize=(7,5))
plt.scatter(df['Screen Size (inches)'], df['Price'], color='green')
plt.title("Screen Size vs Price")
plt.xlabel("Screen Size (inches)")
plt.ylabel("Price (₹)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 9. Graphically show the price distribution of laptops.

price = df['Price']
# price

plt.figure(figsize=(8,5))
plt.hist(price, bins=8, color='orange', edgecolor='black')
plt.title("Price Distribution of Laptops")
plt.xlabel("Price (₹)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#10. Graphically show the contribution of RAM types and explode the RAM having lowest contribution.

ram_counts = df['RAM Type'].value_counts()
labels = ram_counts.index
sizes = ram_counts.values
explode = [0.15 if i == np.argmin(sizes) else 0 for i in range(len(sizes))]

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', startangle=140)
plt.title("Contribution of RAM Types (lowest exploded)")
plt.axis('equal')
plt.show()
```

```python
# 6. Represent categorical data of laptops graphically for laptop Category.

category_count = df['Category'].value_counts()
category_count

plt.figure(figsize = (8,5))
plt.bar(category_count.index, category_count.values, color='skyblue')
plt.title("Laptop count per category")
plt.xlabel("category")
plt.ylabel("Count")
#plt.xticks(rotation=45)
#plt.tight_layout()
plt.show()
```

```python
# 7. Graphically show Influence of Category Class on laptop price.

# Select relevant columns
category_price_data = df[['Category', 'Price']].dropna()

# Get unique categories
categories = category_price_data['Category'].unique()

# Prepare list of prices for each category
price_data = [category_price_data[category_price_data['Category'] == cat]['Price'] for cat in categories]


plt.figure(figsize=(8,5))
plt.boxplot(price_data, labels=categories)
plt.title("Influence of Laptop Category on Price")
plt.xlabel("Category")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 8. Graphically show brand wise laptops count.

brand_wise_laptop_count = df['Brand'].value_counts();
brand_wise_laptop_count

plt.figure(figsize = (8,5))
plt.barh(brand_wise_laptop_count.index, brand_wise_laptop_count.values, color='skyblue')
plt.title("Brand wise laptop count (horizontal Bar chart)")
plt.xlabel("Count")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()
```

```python
# 9. Graphically show the frequency distribution of Operating System.

os_counts = df['Operating System'].value_counts()
os_counts

# plt.figure(figsize = (8, 5))
# plt.bar(os_counts.index, os_counts.values, color='skyblue')
# plt.title("Frequency distribution of Operating System")
# plt.xlabel("Operating System")
# plt.ylabel("Count")
# plt.show()


# Define colors (you can choose any colors)
colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'violet', 'yellow']

plt.figure(figsize = (8, 5))
plt.pie(os_counts, labels = os_counts.index, autopct = '%1.1f%%',  colors=colors[:len(os_counts)], startangle = 140,  shadow= True)
plt.title("Frequency distribution of Operating System")
plt.show()
```

```python
# 10. Graphically show the contribution of all brands and explode the brand having highest contribution.

# Count of laptops per brand
brand_counts = df['Brand'].value_counts()

# # Plot horizontal bar chart
# plt.figure(figsize=(8,5))
# plt.barh(brand_counts.index, brand_counts.values, color='mediumslateblue')
# plt.xlabel("Number of Laptops")
# plt.ylabel("Brand")
# plt.title("Brand Contribution (Horizontal Bar Chart)")
# plt.tight_layout()
# plt.show()

import matplotlib.pyplot as plt


# Find the brand with the highest count
explode = [0.1 if brand == brand_counts.idxmax() else 0 for brand in brand_counts.index]

# Define colors
colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'violet', 'yellow', 'lightcoral', 'lightgrey']

# Plot pie chart with exploded slice for highest brand
plt.figure(figsize=(7,7))
plt.pie(
    brand_counts,
    labels=brand_counts.index,
    autopct='%1.1f%%',
    explode=explode,
    startangle=90,
    shadow=True,
    colors=colors[:len(brand_counts)]
)
plt.title("Brand Contribution (Highlighting Highest)")
plt.show()
```
