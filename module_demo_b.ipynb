# Python Module 
A Python module is a file containing Python definitions and statements. A module can define functions, classes, and variables. A module can also include runnable code. Grouping related code into a module makes the code easier to understand and use. It also makes the code logically organized.

```python
# Create a module file with .py extension
```

```python
# import module
import module_demo
print(module_demo.add(2,3))
print(module_demo.sub(5,3))
```

```python
# renaming or alias of module
import module_demo as md
print(md.add(5,10))
```

# Python built-in modules

```python
# math-- sqrt(), factorial(), sin(), cos(), tan(), degrees(),radians()
import math
print(math.sqrt(25))
print(math.factorial(5))
print(math.sin(25))
print(math.cos(30))
print(math.tan(60))
print(math.degrees(30))
print(math.radians(1718.8733853924696))
```

```python
# datetime -- date(), today(), time(),  strftime()
from datetime import date
today=date.today()
print(today)
print(today.month)
print(today.day)
print(today.year)
```

```python
import datetime
x=datetime.datetime.now()
print(x)
print(x.strftime("%A"))
print(x.strftime("%a"))
print(x.strftime("%B"))
print(x.strftime("%b"))
print(x.strftime("%S"))
print(x.strftime("%H"))
print(x.strftime("%M"))
```

```python
# random -- randint(), random(), choice()
import random
print(random.random())
print(random.randint(1,10))
list=[5,10,15,25.50,15.75,"Hello","Python"]
print(random.choice(list))
```
