```python
# Modes
# "r" - reading
# "a" - appending
# "w" - writing
# "x" - creating

# "t" - text file
# "b" - binary file
```

```python
# File handling functions
# open() & read()
f=open("file2.txt","r")
print(f.read())

```

```python
f=open("file2.txt","r")
print(f.read(10))
```

```python
f=open("file2.txt","r")
print(f.readline())
```

```python
f=open("file2.txt","r")
for i in f:
    print(i)
```

```python
# close()
f=open("file2.txt","r")
f.close()
#f=open("file2.txt","r")
print(f.read())
```

```python
# appending with write()
f=open("file2.txt","a")
f.write("Python.... Python...")
f.close()
f=open("file2.txt","r")
print(f.read())
```

```python
# overwrite with write()
f=open("file2.txt","w")
f.write("New Contents....")
f.close()
f=open("file2.txt","r")
print(f.read())
```

```python
# deleting a file
f.close()
import os
os.remove("file2.txt")
```
