```python
import pandas as pd
```

```python
d1={"Name":["Manish","Bharat","Rahul","Akash"], "Age":[22,25,24,28]}
d2={"Name":["Rajesh","Bharat","Rahul","Ravi"], "Age":[23,25,24,27]}
d1
```

```python
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
df2
```

```python
pd.merge(df1,df2, how="left", on="Name")
```

```python
pd.merge(df1,df2, how="right", on="Name")
```

```python
pd.merge(df1,df2, how="inner", on="Name")
```

```python
pd.merge(df1,df2, how="outer", on="Name")
```
