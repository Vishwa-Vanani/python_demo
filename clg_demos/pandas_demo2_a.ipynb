```python
import pandas as pd
```

```python
d1={"Name":["Manish","Bharat","Rahul","Aakash"],"Age":[22,25,27,28]}
d2={"Name":["Rajesh","Bharat","Rahul","Ravi"],"Age":[23,25,27,26]}
d2
```

```python
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
print(df1)
print(df2)
```

```python
pd.merge(df1,df2, how="left", on="Name")
```

```python
pd.merge(df1,df2, how="right", on="Name")
```

```python
pd.merge(df1,df2, how="inner", on="Name")
```

```python
pd.merge(df1,df2, how="outer", on="Name")
```
