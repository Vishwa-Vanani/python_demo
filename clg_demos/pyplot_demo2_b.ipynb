```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df.head()
```

```python
age=df['Age']
age.head()
```

```python
fare=df['Fare']
fare.head()
```

```python
plt.scatter(age,fare,c='red',marker='<')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
df1=df['Survived'].value_counts()
df1
```

```python
plt.bar(df1.index,df1.values)
plt.show()
```

```python
df2=df['Pclass'].value_counts()
df2
```

```python
plt.bar(df2.index,df2.values)
plt.xlabel('Category')
plt.ylabel('Value Counts')
plt.show()
```

```python
plt.barh(df2.index,df2.values)
plt.xlabel('Value Counts')
plt.ylabel('Category')
plt.show()
```

```python
sd=df.groupby('Pclass')['Survived'].value_counts().unstack()
#sd.plot(kind='bar', stacked=False)
sd.plot(kind='barh', stacked=True)
plt.xlabel('Survived counts')
plt.ylabel('Survived data')
plt.title('Survival Counts by Passenger Class')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
